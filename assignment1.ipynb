```python
arr_size=5
student_info=[]

for i in range(arr_size):
   student_name=input(f"Enter Your Name {i+1}: ")
   student_grade=int(input(f"Enter Your Grade {i+1}: "))
   student_info.append([student_name,student_grade])

print("All Student:",student_info)
Higher_name=student_info[0][0]
Higher_Grade=student_info[0][1]
Lower_name=student_info[0][0]
Lower_Grade=student_info[0][1]
for student_name,student_grade in student_info:
  if student_grade>Higher_Grade:
    Higher_Grade=student_grade
    Higher_name=student_name
  elif student_grade<Lower_Grade:
    Lower_Grade=student_grade
    Lower_name=student_name
print("Higher student Name:",Higher_name,"Got",Higher_Grade)
print("Lower student Name:",Lower_name,"Got",Lower_Grade)

```

```python
size=3
person_info=[]
for i in range(size):
  name=input("Enter Your Name: ")
  age=int(input("Enter Your Age: "))
  person_info.append([name,age])
print("All persons:",person_info)


for name,age in person_info:
  if age<18:
    stats="Young"
  elif age<=60:
    stats="Adult"
  else:
    stats="Senior"
  print(f"{name} {age} Years Old: {stats}")
```

```python
size=10
temp=[]
for i in range(size):
  value=int(input(f"Enter Temperature {i+1}: "))
  temp.append(value)
print(temp)
count_above30=0
count_below15=0
count_between15_30=0
for value in temp:
  if value>=30:
    count_above30=count_above30+1
  elif value<=15:
    count_below15=count_below15+1
  else:
    count_between15_30=count_between15_30+1
print("Your have ",count_above30," Temperature above 30")
print("Your have ",count_below15," Temperature below 15")
print("Your have ",count_between15_30," Temperature between 15-30")
```

```python
List=[]
while True:
  print("=============")
  print("1-Add item ")
  print("2-Remove item by name")
  print("3-Show sorted list")
  print("4-Show List")
  print("5-Exit")
  print("=============")
  choice=int(input("Enter Your Choice: "))


  if choice==1:
    item=input("Enter Item: ")
    if item:
      List.append(item)
      print("Item Added Succsufly")
    else:
      print("Item Name Empty Try Again")

  elif choice==2:
    item=input("Enter Item: ")
    if item in List:
      List.remove(item)
      print("Item Removed Succsufly")
    else:
      print(item,"Cant Found on the List")
  elif choice==3:
    List.sort()
    print("Sorted List:",List)
  elif choice==4:
    print("List:",List)
  elif choice==5:
    print("Bye Bye")
    break
  else:
    print("Invalid Choice")


```
